```python
import pandas as pd

# Load your data
data = pd.read_csv('wine_production.csv')
data.fillna('0', inplace=True)
data.set_index('Year', inplace=True)
data = data.loc[1900:]
for column in data.columns:
    data[column] = pd.to_numeric(data[column].str.replace(',', ''), errors='coerce').fillna(0).astype(int)
data = data.T
data = data.sort_index()

country_info = pd.read_csv('country_info.csv')
country_info.set_index('Country Name', inplace=True)

combined_data = country_info.merge(data, left_index=True, right_index=True)
```

```python
continent_data = combined_data.drop(columns='Image URL')
continent_data = continent_data.groupby('Region').sum()
continent_data.to_csv('continent_wine_production.csv')
```

```python
production = combined_data.drop(columns=['Region', 'Image URL']).sum(axis=1)

# Get the top 10 producers
top_producers = production.sort_values(ascending=False).head(7)
data_top_producers = combined_data.loc[top_producers.index]

# Drop the columns again for the top producers data
data_top_producers = data_top_producers.drop(columns=['Region', 'Image URL'])

# Calculate the production for the rest of the countries
rest_producers = combined_data.drop(index=top_producers.index)
rest_production = rest_producers.drop(columns=['Region', 'Image URL']).sum(axis=0)

# Create a row for the rest of the countries
other_countries_row = pd.DataFrame([rest_production], columns=data_top_producers.columns, index=["Other countries"])

# Append the 'Other countries' row to the data of top producers
top_countries = pd.concat([data_top_producers, other_countries_row])
top_countries.to_csv('top_producers.csv')
```
